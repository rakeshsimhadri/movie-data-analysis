- Treat Outliers

```python
#Calculate Q1

Q1=df["Rating"].quantile(0.25)
print("Q1:",Q1)


Q2=df["Rating"].quantile(0.5)
print("Q2:",Q2)


Q3=df["Rating"].quantile(0.75)
print("Q3:",Q3)


#Calculate IQR 

IQR=Q3-Q1
print("IQR:",IQR)


#Calculate Lower Limit of outliers

Lower_Limit=Q1-(1.5*IQR)
print("Lower_Limit:",Lower_Limit)


Upper_Limit=Q3+(1.5*IQR)
print("Upper_Limit:",Upper_Limit)
```

```python
df[(df["Rating"]<Lower_Limit) | (df["Rating"]>Upper_Limit)]
```

- Movies that earn high revenue usually have good ratings and a lot of audience votes.

The movie with the highest revenue (~936M) has about 661,608 votes and a rating of 8.1.

Movies with ratings above 7.5 tend to earn more money, showing that when people like the movie, it performs well in theaters.

Also, movies with many votes mean more people watched and talked about them, which helps them earn more money.

Revenue increases when more people watch the movie (high votes). However, when the rating drops below 7.0, the revenue also drops because fewer people recommend or re-watch the movie. This shows that both audience reach (Votes) and content quality (Rating) together influence revenue.

```python
bins = [4,5,6,7,8,9]
labels = ['4 Star','5 Star','6 Star','7 Star','8 Star']
df['Rating_Group'] = pd.cut(df['Rating'], bins=bins, labels=labels, include_lowest=True)

df.groupby(["Year","Rating_Group"],observed=True)["Revenue (Millions)"].agg(["max","count"])
```

- This chart shows how movie revenue varies each year across different rating groups (4â˜… to 8â˜…).
Movies with higher ratings (7â˜… and 8â˜…) consistently achieved higher maximum revenue across the years.
Years like 2008, 2009, 2013, and 2015 show clear revenue peaks, especially in the 7â˜… and 8â˜… categories, indicating blockbuster releases during those years.
Movies with lower ratings (4â˜… and 5â˜…) show much lower revenue, meaning these movies rarely become commercial hits.

Conclusion: Higher-rated movies generally perform better at the box office. Rating is an important factor influencing revenue
